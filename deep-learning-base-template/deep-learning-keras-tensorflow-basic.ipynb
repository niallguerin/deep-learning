##### Machine Learning Keras Neural Network Construction and Compilation
Now we create a neural network using Keras. We create a sequential model and add dense layers. The input shape is very important as this is related to the number of valid feature columns in the Iris dataset used to train the model(4).

For information on relu, sigmoid, tanh, and softmax activation functions, refer to the web references section at the end of this notebook.

The Keras neural network structure is as follows based on trial and error tests over a short period of 1-2 hours:
- 27 nodes on the first hidden input layer and an input shape of 4 to represent our 4 feature columns in iris dataset
- 18 nodes on the second layer using the rectified linear (relu) activation function
- 3 nodes on the third and final output layer which uses the softmax function - 3 nodes based on one-hot-encoding for the 3 categorical labels for our target Species flower variety

Finally, we add a plot statement to visually plot the structure of the neural network layers and number of neurons per layer.

```python
# define the keras model and number of layers: 99-100% accuracy in predictions on training and 93-96% on test
keras_tf_model = Sequential()
keras_tf_model.add(Dense(27, input_shape=(4,), activation='relu'))
keras_tf_model.add(Dense(18, activation='relu'))
keras_tf_model.add(Dense(3, activation='softmax'))
plot_model(keras_tf_model, to_file='model_plot_iris.png', show_shapes=True, show_layer_names=True)
```

Now we need to compile our Keras neural network model. This is the step when we utilize the numeric library horsepower under the hood. In our case, we will be using TensorFlow. The goal here in the training phase is to find the optimal weights that will enable this pretty simple network model to map an input dataset observation to an output label i.e. one of three Species of flower. We need to do the following at this point:

- specify a loss function for optimal weight calculations in our neural network model
- specify an optimizer that searches through different weights for our neural network
- specify any metrics for data collection during training of our neural network

```python
# compile the keras model - NB - ensure you use categorical NOT binary here as dataset labels now 3 instead of 2
keras_tf_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

The above line of code does the three things set out in our objective for this step:

- categorical_crossentropy represents our loss function needed for multi-class classification problems like iris
- adam is the optimizer that searches through different weights. It is a version of gradient descent. For more information refer to this article and paper: https://arxiv.org/abs/1412.6980
- metrics with 'accuracy' is used to track classification accuracy of our neural network model during training

```python
keras_tf_model.summary()
```

##### Machine Learning Keras Neural Network Model Training
The next step involves fitting the neural network model above to our Iris training dataset defined earlier. We need to consider two additional parameters at this step:
- number of Batches (number of dataset records processed before weights updated within an epoch)
- number of Epochs (number of iterations over the ENTIRE dataset used for learning, comprised of 1 or many batches)

```python
# fit the keras model on the dataset - I have adjusted epochs and batch size below based on my own testing on laptop
# epochs and batch_size is a trial and error repeat test and record exercise
history = keras_tf_model.fit(X_train, y_train, epochs=500, batch_size=15)
```

##### Evaluate Performance on Training Dataset
We will use below plots to access the history values for loss and accuracy to evaluate performance on the training dataset.

```python
# plot model loss in training
plt.plot(history.history['loss'], color="Orange")
plt.title('Keras-TensorFlow Neural Network model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training Data'], loc='upper right')
plt.show()
```

```python
# plot history for model accuracy in training
plt.plot(history.history['accuracy'], color="Blue")
plt.title('Keras-TensorFlow Neural Network model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training Data'], loc='upper left')
plt.show()
```

##### Evaluate Performance on Training Dataset
Use below function format per tutorial to print loss and accuracy performance metrics from the model training. Credit the source as a comment in the cell and add link to main Web References at end of this template.

```python
# evaluate the keras model on training dataset - see format on machinelearningmastery.com
# https://machinelearningmastery.com/display-deep-learning-model-training-history-in-keras/
_, accuracy = keras_tf_model.evaluate(X_train, y_train)
print('Training Data Accuracy: %.2f' % (accuracy*100))
```
